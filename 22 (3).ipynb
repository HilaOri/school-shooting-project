##### A closer look at the months during Covid years

```python
start_date = "2020-01-01"
filtered_df = sm_df[sm_df['Date'] >= start_date]
filtered_df['YearMonth'] = filtered_df['Date'].dt.to_period('M')
events_per_month = filtered_df.groupby('YearMonth').size().reset_index(name='Count')
events_per_month['YearMonth'] = events_per_month['YearMonth'].dt.to_timestamp()

highlight_periods = [
    ('2020-01', '2020-07', 'First + Second Wave'),
    ('2020-10', '2021-06', 'Third Wave'),
    ('2021-10', '2021-10', 'Fourth Wave')
]

plt.figure(figsize=(12, 6))
sns.lineplot(x='YearMonth', y='Count', data=events_per_month, marker='o', color='purple')
plt.title('Number of Events per Month from January 2020 (Linear Scale)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Events', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)

for start, end, label in highlight_periods:
    plt.axvspan(start, end, color='green', alpha=0.2)
    plt.axvline(pd.to_datetime(start), color='black', linestyle='--', linewidth=1)
    plt.text(pd.to_datetime(start), plt.ylim()[1] * 0.95, label, rotation=90, verticalalignment='top', color='black', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

```

##### Barplot depicting how many cases there were by time period

##### Number of shooting cases by Psychological Factors and Time Period

```python
sm_df = sm_df[sm_df['Time_Period'] != 'Not A School Day']
factors = ['Hostages', 'Barricade', 'Officer_Involved', 'Bullied', 'Domestic_Violence', 'Gang_Related', 'Preplanned']
time_periods = sm_df['Time_Period'].unique()

fig, axes = plt.subplots(nrows=(len(factors) + 1) // 2, ncols=2, figsize=(15, 20), constrained_layout=True)
axes = axes.flatten()

for i, factor in enumerate(factors):
    filtered_df = sm_df[sm_df[factor] == 'Yes']
    
    count_data = filtered_df['Time_Period'].value_counts().reindex(time_periods, fill_value=0).sort_values(ascending=False)
    
    sns.barplot(x=count_data.index, y=count_data.values, ax=axes[i], palette='viridis')
    axes[i].set_title(f'{factor}')
    axes[i].set_xlabel('Time Period')
    axes[i].set_ylabel('Number of Events')
    axes[i].tick_params(axis='x', rotation=45)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
plt.suptitle('Number of Shooting Cases by Psychological Factors and Time Period', fontsize=16)
plt.show()
```

#####  Heatmap of number of events by psychological factors and time period

```python
sm_df = sm_df[sm_df['Time_Period'] != 'Not A School Day']

factors = ['Hostages', 'Barricade', 'Officer_Involved', 'Bullied', 'Domestic_Violence', 'Gang_Related', 'Preplanned']
time_periods = sm_df['Time_Period'].dropna().unique() 

heatmap_data = pd.DataFrame(index=time_periods, columns=factors).fillna(0)

for factor in factors:
    for period in time_periods:
        heatmap_data.loc[period, factor] = sm_df[(sm_df['Time_Period'] == period) & (sm_df[factor] == 'Yes')].shape[0]

heatmap_data = heatmap_data.fillna(0).astype(int)

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data.T, annot=True, fmt='d', cmap='viridis', linewidths=.5, cbar_kws={'label': 'Number of Events'})
plt.title('Number of Events by Psychological Factors and Time Period')
plt.xlabel('Time Period')
plt.ylabel('Factors')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

From this heatmap it can be seen that the strongest connection is between the sports events and the gang related cases. 
In addition, it can be seen that most all of the preplanned cases, were durning morning classes.

```python
sport_police = sm_df[sm_df['Time_Period'] == 'Sport Event']
sport_police_count = sport_police[sport_police['Officer_Involved'] == 'Yes'].shape[0]
print("The number of officer involved in sport events:", sport_police_count)

gang_related_count = sport_police[sport_police['Gang_Related'] == 'Yes'].shape[0]
print("The number of gang related events in sport events:", gang_related_count)
```

```python
#sm_df["First_Shot"] = sm_df["First_Shot"].apply(mixed_to_time)

specific_time = datetime.time(15, 0, 0)
filtered_df = sm_df[sm_df["First_Shot"] == specific_time]
```

##### Value count in which location there were cases at 15:00:00

```python
filtered_df['Location'].value_counts()
```

```python
leaving_school = ["Parking Lot", "School Bus", "Outside on School Property", "Front of School", "Parking Lot (Bus)"]
in_school = ["Beside Building", "Bathroom", "Hallway", "Football Field/Track", "Classroom", "Field (General)", "Gym"]
```

##### Number of Events by Location at 15:00

```python
location_counts = filtered_df["Location"].value_counts().reset_index()
location_counts.columns = ['Location', 'Counts']
my_order = location_counts['Location']

plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.barplot(data=location_counts, x='Location', y='Counts', palette='viridis', order=my_order)

plt.title(f'Number of Events by Location at {specific_time}')
plt.xlabel('Location')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.show()
```

##### Percentage of Events by Location Category at 15:00

```python
filtered_df['Location_Category'] = np.where(filtered_df['Location'].isin(leaving_school), 
                                            'Leaving School', 
                                            np.where(filtered_df['Location'].isin(in_school),
                                                     'In School',
                                                     'Other'))

location_counts = filtered_df['Location_Category'].value_counts().reset_index()
location_counts.columns = ['Location_Category', 'Counts']

total_counts = location_counts['Counts'].sum()
location_counts['Percentage'] = (location_counts['Counts'] / total_counts) * 100

plt.figure(figsize=(8, 8))
colors = ['#568203', '#78866B', '#D3D3D3']  # צבעים עבור כל קטגוריה
plt.pie(location_counts['Percentage'], labels=location_counts['Location_Category'], autopct='%1.1f%%', colors=colors)

plt.title(f'Percentage of Events by Location Category at {specific_time}')
plt.show()
```
